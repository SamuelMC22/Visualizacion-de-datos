```python
import pandas as pd

df = pd.read_csv('/content/spotify_churn_dataset.csv')
```

```python
df.head()
```

```python
display(df.head())
display(df.info())
```

```python
display(df.describe())
display(df.describe(include='all'))
```

```python
valores_faltantes = df.isnull().sum()
print(valores_faltantes)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.histplot(df['age'], bins=20)
plt.title('Distribución de edad')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.show()

plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='gender')
plt.title('Cantidad de usuarios por genero')
plt.xlabel('Genero')
plt.ylabel('cantidad')
plt.show()

plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='subscription_type')
plt.title('Cantidad de usuario por tipo de suscripción')
plt.xlabel('Suscripcion')
plt.ylabel('Cantidad')
plt.show()

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='listening_time', y='songs_played_per_day', alpha=0.5)
plt.title('Tiempo escuchado vs canciones por dia')
plt.xlabel('Tiempo escuchado (horas)')
plt.ylabel('Canciones escuchadas por dia')
plt.show()

plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='device_type', y='skip_rate')
plt.title('Tasa de saltadas media por tipo de dispositivo')
plt.xlabel('Tipo de dispositivo')
plt.ylabel('Tasa de saltadas media')
plt.show()
```

Entre las variables se encuentran user\_id, gender, age, country, subscription\_type, listening\_time, songs\_played\_per\_day, skip\_rate, device\_type, ads\_listened\_per\_week, offline\_listening e is\_churned. En la exploración inicial se visualizaron las primeras filas, se revisaron los tipos de datos y se obtuvo información general del conjunto. Posteriormente, se generaron estadísticas descriptivas para variables numéricas y no numéricas, lo que permitió identificar distribuciones y conteos relevantes. Finalmente, se realizaron visualizaciones como un histograma de age, gráficos de barras para gender y subscription\_type, un diagrama de dispersión entre listening\_time y songs\_played\_per\_day y un gráfico de barras del promedio de skip\_rate por device\_type
